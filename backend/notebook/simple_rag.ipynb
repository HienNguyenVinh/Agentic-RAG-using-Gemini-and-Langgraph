```python
from sentence_transformers import SentenceTransformer
import numpy as np

model = SentenceTransformer('hiieu/halong_embedding', trust_remote_code=True)

def get_embedding(text:str) -> list[float]:
    if not text.strip():
        print('Attemped to get embedding for empty text.')
        return []

    vector = model.encode(text)
    return np.ndarray.tolist(vector)
```

```python
import chromadb

chroma_client = chromadb.PersistentClient(path="/content/chroma_db")

collection = chroma_client.get_or_create_collection(name="book_collection")
```

```python
import numpy as np

df['embedding_vector'] = df['embedding_vector'].apply(eval)
df['embedding_vector'] = df['embedding_vector'].apply(np.array)
```

```python
for i, row in df.iterrows():
    collection.add(
        ids=[str(i)],
        embeddings=[row['embedding_vector']],
        metadatas={
            'name':row['name'],
            'author':row['author'],
            'category':row['category'],
            'description':row['description'],
            'price':row['price'],
            'summarize':row['summarize'],
        }
    )

print('Dữ liệu đã lưu vào Chroma')
```

```python
query = "Sách về những con mèo"

query_embedding = get_embedding(query)

results = collection.query(
    query_embeddings=query_embedding,
    n_results=3
)

results
```

```python
from rank_bm25 import BM25Okapi
from underthesea import word_tokenize

tokenized_corpus = [word_tokenize(doc.lower()) for doc in df['summarize']]

bm25 = BM25Okapi(tokenized_corpus)
```

```python
def hybrid_search(query, bm25_weight=0.5, k = 5):
    tokenized_query = word_tokenize(query.lower())
    bm25_scores = bm25.get_scores(tokenized_query)
    bm25_top_k = np.argsort(bm25_scores)[::-1][:k]

    embedding_query = get_embedding(query)
    vector_scores = collection.query(
        query_embeddings=[embedding_query],
        n_results=k
    )

    hybrid_scores = {}

    bm25_max = max(bm25_scores) if max(bm25_scores) > 0 else 1
    for idx in bm25_top_k:
        hybrid_scores[idx] = (bm25_scores[idx] / bm25_max) * bm25_weight

    for i, doc in enumerate(vector_scores["ids"][0]):
        idx = int(doc)
        score = (1 - i / k) * (1 - bm25_weight)
        hybrid_scores[idx] = hybrid_scores.get(idx, 0) + score


    sorted_results = sorted(hybrid_scores.items(), key=lambda x: x[1], reverse=True)

    results = []
    for idx, score in sorted_results[:k]:
        row = df.iloc[idx]
        results.append({
            "name": row["name"],
            "author": row["author"],
            "category": row["category"],
            # "description": row["description"],
            "highlight": row["highlight"],
            "price": row["price"],
            # "full_description": row["full_description"],
            # "description_length": row["description_length"],
            "summarize": row["summarize"],
            "score": round(score, 2)
        })

    return results
```

```python
query = "Sách về những con mèo"

results = hybrid_search(query, k=3)
results
```

```python
def get_result(query):
    results = hybrid_search(query, k=3)

    full_result = ""
    for res in results:
        if res.get('summarize'):
            full_result += res['summarize']
        if res.get('price'):
            full_result += f"\nGiá: {res.get('price')}" + " VND.\n"
        else:
            full_result += "Liên hệ đế trao đổi chi tiết."

    return full_result
```

```python
query = "Sách về những con mèo"
source_information = get_result(query)

source_information
```
